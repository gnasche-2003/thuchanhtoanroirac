#a. Chng trnh (module) vit theo k thut lp (khng  quy). 

```python
def fibo(n):
  f0 = 0
  f1 = 1
  fn = 1

  if (n == 0) or (n == 1):
    return 
  else:
    for i in range(2,n):
      f0 = f1
      f1 = fn
      fn = f1 + f0
    return fn

n = int(input("Nhap vao so n: "))
print(fibo(n))
```

# b. Chng trnh (module) vit bng cch tnh ton trc tip gi tr s Fibonacci theo cng 
thc ma trn (s hc trong thc hnh i s tuyn tnh). 


```python
def fibo_mem(n,so_da_tinh = {0:0,1:1}):
  if n not in so_da_tinh:
    so_da_tinh[n] = fibo_mem(n-1, so_da_tinh) + fibo_mem(n-2, so_da_tinh)
  return so_da_tinh[n]
```

```python
fibo_mem(3)
```
