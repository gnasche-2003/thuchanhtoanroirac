#a. Chương trình (module) viết theo kỹ thuật lặp (không đệ quy). 

```python
def fibo(n):
  f0 = 0
  f1 = 1
  fn = 1

  if (n == 0) or (n == 1):
    return 
  else:
    for i in range(2,n):
      f0 = f1
      f1 = fn
      fn = f1 + f0
    return fn

n = int(input("Nhap vao so n: "))
print(fibo(n))
```

# b. Chương trình (module) viết bằng cách tính toán trực tiếp giá trị số Fibonacci theo công 
thức ma trận (sẽ học trong thực hành đại số tuyến tính). 


```python
def fibo_mem(n,so_da_tinh = {0:0,1:1}):
  if n not in so_da_tinh:
    so_da_tinh[n] = fibo_mem(n-1, so_da_tinh) + fibo_mem(n-2, so_da_tinh)
  return so_da_tinh[n]
```

```python
def fibona(n): 
 bat_dau = datetime.now() # Ghi nhận thời điểm bắt đầu
 ketqua = fibo(n)
 for i in range(n+1): 
  ketqua = ketqua + i 
 thoi_gian = str(milisec_passed(bat_dau)) # Tính toán thời gian trôi qua
 print(thoi_gian) 
 return ketqua 
print(fibona(10)) 
```
